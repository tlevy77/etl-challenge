### Cleaning First Two Tables

```python
times = df2['length']
new_len_col = []

for x in range(len(times)):
    le = times[x].split(":")
    if len(le)==2:
        h = "0"
        m = le[0]
        s = le[1]
    else:
        h = le[0]
        m = le[1]
        s = le[2]
    new_len_col.append(h.rjust(2, "0") + ":" + m.rjust(2, "0") + ":" + s.rjust(2, "0"))
```

```python
import datetime
test_dates = df2['date']
test_format = "%Y-%m-%d"
new_date_column = []
for x in range(len(test_dates)):
    try:
        dt2 = datetime.datetime.strptime(test_dates[x], test_format).date()
        new_date_column.append(dt2)
    except ValueError:
        new_date_column.append(0)
        
```

```python
new_df1 = df1[['locId','comName','sciName','lat','lng','subId']].copy()
new_df1 = new_df1.rename(columns={"locId":"locid","comName":"comname","sciName":"sciname","subId":"subid"})
```

```python
new_df2 = df2[['id','gen','sp','en','lat','lng']].copy()
new_df2['len'] = new_len_col
new_df2['d'] = new_date_column
```

```python
new_df2 = new_df2[new_df2.d != 0]
```

```python
new_df1 = new_df1.drop_duplicates().copy()
new_df2 = new_df2.drop_duplicates().copy()
```

### Web Scraping Audubon Bird Guide ETL

```python
import os
import requests
from bs4 import BeautifulSoup as bs
import pymongo
from splinter import Browser
from bs4 import BeautifulSoup
from webdriver_manager.chrome import ChromeDriverManager
```

```python
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
executable_path = {'executable_path': '/usr/local/bin/chromedriver'}
```

```python
url = 'https://www.audubon.org/bird-guide'
browser.visit(url)
```

```python
com_names = []
sci_names = []

for x in range(0, 26):
    html = browser.html
    soup = BeautifulSoup(html, 'html.parser')
    divs = soup.find_all('div', class_="page-"+str(x)+" views-row columns tiny-6 medium-4 large-3")
    
    for div in divs:
        try:
            h3 = div.find('h3', class_ = 'common-name custom-h4').text
            p = div.find('p', class_ = 'scientific-name').text.strip()
            com_names.append(h3)
            sci_names.append(p)
        except:
            pass
    
    if x < 25:
        browser.visit(url + "?page=" + str(x+1) + "&amp;site=vt&amp;nid=12608")
    else:
        print("Scraping Completed!")
    
    
```

```python
browser.quit()
```

```python
import pandas as pd

species2_df = pd.DataFrame({"comname":com_names,"sciname":sci_names})
#species2_df.to_csv('species2.csv', index=False)
species2_df
```
